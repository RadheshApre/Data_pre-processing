```python
#Train-Test Split + Feature Scaling (Best Practice)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

data = {
    'Age': [22, 25, 47, 52, 46, 56, 48],
    'Salary': [25000, 27000, 55000, 48000, 52000, 60000, 58000],
    'Purchased': [0, 0, 1, 1, 1, 0, 1]
}
df = pd.DataFrame(data)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)

print("X_train_scaled:\n", X_train_scaled)
print("\nX_test_scaled:\n", X_test_scaled)
```

```python
#Following is the dataset with columns Country, Age, Salary, and Purchased. You apply a ColumnTransformer that uses OneHotEncoder on Country 
#and StandardScaler on Age and Salary, then train a Logistic Regression model.
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression

# Step 1: Create dataset
data = {
    'Country': ['France', 'Spain', 'Germany', 'Spain', 'France', 'Germany', 'France'],
    'Age': [22, 25, 47, 52, 46, 56, 48],
    'Salary': [25000, 27000, 55000, 48000, 52000, 60000, 58000],
    'Purchased': [0, 0, 1, 1, 1, 0, 1]
}
df = pd.DataFrame(data)

# Step 2: Split features and target
X = df.iloc[:, :-1]   # Country, Age, Salary
y = df.iloc[:, -1]    # Purchased

# Step 3: Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)

# Step 4: Preprocessing (Encoding + Scaling)
ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), ['Country']),       # Encode categorical
        ('scaler', StandardScaler(), ['Age', 'Salary']) # Scale numerical
    ]
)

X_train_processed = ct.fit_transform(X_train)
X_test_processed = ct.transform(X_test)

# Step 5: Train model
model = LogisticRegression()
model.fit(X_train_processed, y_train)

# Step 6: Evaluate
print("Accuracy:", model.score(X_test_processed, y_test))
```
